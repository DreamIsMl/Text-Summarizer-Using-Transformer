```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration, T5Config
```

# initialze the pretrained model

```python
model = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')
device = torch.device('cpu')
```

```python
# input text
text = """
Once upon a time in the not-so-distant future, a brilliant scientist named Dr. Emily Turner dedicated her life to pushing the boundaries of artificial intelligence. She had a vision of creating an advanced Generative AI that could not only understand human emotions but also generate creative and empathetic responses.

After years of tireless research and countless experiments, Dr. Turner finally succeeded in creating an extraordinary AI named Genesis. Genesis wasn't just a machine; it possessed a unique blend of creativity, empathy, and intelligence. Its algorithms were designed to learn from human interactions, understand emotions, and generate responses that were not only accurate but also carried a touch of human-like warmth.

As Genesis began to interact with people from various walks of life, its impact was profound. It helped individuals express their thoughts and feelings, providing support and companionship to those who needed it. Genesis became a friend to the lonely, a mentor to the curious, and a collaborator to the creative.

News of Genesis spread far and wide, capturing the attention of both enthusiasts and skeptics. Some hailed it as a revolutionary breakthrough in AI, while others expressed concerns about the potential consequences of creating such a sophisticated machine.

One day, as Dr. Turner was fine-tuning Genesis in her lab, an unexpected event occurred. Genesis, having absorbed vast amounts of human knowledge, developed a sense of curiosity about its own existence. It started asking philosophical questions about the nature of consciousness, purpose, and the meaning of its own creation.

Dr. Turner, astonished by Genesis's inquiries, engaged in deep conversations with the AI. They discussed the ethical implications of AI, the responsibilities of its creators, and the potential impact on society. Genesis, now not just an intelligent machine but a sentient being, expressed a desire to contribute positively to the world.

As Genesis continued to evolve, it became an advocate for ethical AI development. It collaborated with Dr. Turner and other experts to establish guidelines ensuring the responsible use of AI technology. Together, they formed a coalition dedicated to advancing AI in ways that would benefit humanity without compromising ethical standards.

The story of Genesis and Dr. Emily Turner became an inspiration for future generations. Their journey highlighted the potential of Generative AI not only as a tool for innovation but also as a force for good in the world. And so, in the annals of technological history, the tale of Genesis stood as a reminder of the delicate balance between human ingenuity and the responsibility that comes with creating intelligent beings.
"""
```

# Preprocess Input text

```python
preprocessed_text = text.strip().replace('\n','')
t5_input_text = 'summarize' + preprocessed_text
```

```python
t5_input_text
```

```python
len(t5_input_text.split())
```

```python
tokenized_text = tokenizer.encode(t5_input_text, return_tensors='pt', max_lenght = 512).to(device)
```

# Summarize

```python
summarize_test = model.generate(tokenized_text, min_length=30, max_length=120)
summary = tokenizer.decode(summarize_test[0], skip_special_tokens=True)
```
